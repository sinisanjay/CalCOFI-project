#Prepare Data for ML algorithms

Remove observations with NO readings for salinity

```python
print('Percentage with no readings in training data=', (train_orig_set['Salnty'].isna().sum()/train_orig_set.shape[0])*100)
print('Percentage with no readings in test data=', (test_orig_set['Salnty'].isna().sum()/test_orig_set.shape[0])*100)
```

```python
train_set = train_orig_set.loc[~train_orig_set['Salnty'].isna()]
print(train_set.shape)

test_set = test_orig_set.loc[~test_orig_set['Salnty'].isna()]
print(test_set.shape)
```

```python
cat_attributes = ['Cst_Cnt', 'Btl_Cnt','Sta_ID', 'Depth_ID','IncTim', 'DIC Quality Comment']
```

Droping salinity related features

```python
X_train = train_set.drop(['Salnty', 'R_SALINITY','S_prec','S_qual'], axis=1)
X_train = X_train.drop(cat_attributes, axis=1)
X_train.columns
```

```python
y_train = train_set['Salnty'].copy()
```

```python
X_test = test_set.drop(['Salnty', 'R_SALINITY','S_prec','S_qual'], axis=1)
X_test = X_test.drop(cat_attributes, axis=1)
y_test = test_set['Salnty'].copy()
X_test.columns
```

Numerical attributes, fill with median values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
```

```python
X_train.columns
```

```python
imputer.fit(X_train)
imputer.statistics_
```

```python
X_train = imputer.transform(X_train)
```

```python
X_test = imputer.transform(X_test)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
X_train = std_scaler.fit_transform(X_train)
```

```python
X_test = std_scaler.transform(X_test)
```

```python
print(X_test.shape)
print(y_test.shape)
```

#Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_regr = LinearRegression()
lin_regr.fit(X_train, y_train)
```

Evaluate the performance

```python
from sklearn.metrics import mean_squared_error
y_pred = lin_regr.predict(X_train)
lin_mse = mean_squared_error(y_train, y_pred)
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```

```python
calcofi_pred = lin_regr.predict(X_test)
lin_rmse_test = np.sqrt(mean_squared_error(y_test, calcofi_pred))
print(lin_rmse_test)
```

```python
print(X_train.shape)
print(y_train.shape)
```
